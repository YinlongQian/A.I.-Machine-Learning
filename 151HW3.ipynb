```python
mean = [0, 0]
cov = [[9, 0], [0, 1]]
```

```python
import numpy as np
import matplotlib.pyplot as plt
x, y = np.random.multivariate_normal(mean, cov, 100).T
plt.plot(x, y, 'x')
```

```python
plt.axis('equal')
plt.show()
```
