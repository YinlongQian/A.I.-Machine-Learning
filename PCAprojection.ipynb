```python
import numpy as np
file = open('predicate-matrix-continuous.txt', 'r')
data = np.loadtxt(file)
```

```python
file2 = open('classes.txt', 'r')
lines = file2.readlines()
classes= [x.strip()[2:] for x in lines]
for index in range(9,50):
    classes[index] = classes[index][1:]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(data)
reduced
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20, 20

import matplotlib.pyplot as plt
sample = np.array(reduced)
x,y=sample.T
plt.scatter(x,y)
for index in range(50):
    if index != 5 and index != 9:
        plt.annotate(classes[index], xy=(reduced[index][0], reduced[index][1]))
plt.annotate(classes[5], xy=(reduced[5][0], reduced[5][1]),xytext=(reduced[5][0]-12, reduced[5][1]-2))
plt.annotate(classes[9], xy=(reduced[9][0], reduced[9][1]),xytext=(reduced[9][0]-10, reduced[9][1]-2))
plt.show()
```
