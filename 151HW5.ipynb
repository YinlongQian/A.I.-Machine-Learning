```python
from struct import unpack
import numpy as np
import matplotlib.pylab as plt 
import math
import scipy

def loadmnist(imagefile, labelfile):

    # Open the images with gzip in read binary mode
    images = open(imagefile, 'rb')
    labels = open(labelfile, 'rb')

    # Get metadata for images
    images.read(4)  # skip the magic_number
    number_of_images = images.read(4)
    number_of_images = unpack('>I', number_of_images)[0]
    rows = images.read(4)
    rows = unpack('>I', rows)[0]
    cols = images.read(4)
    cols = unpack('>I', cols)[0]

    # Get metadata for labels
    labels.read(4)
    N = labels.read(4)
    N = unpack('>I', N)[0]

    # Get data
    x = np.zeros((N, rows*cols), dtype=np.uint8)  # Initialize numpy array
    y = np.zeros(N, dtype=np.uint8)  # Initialize numpy array
    for i in range(N):
        for j in range(rows*cols):
            tmp_pixel = images.read(1)  # Just a single byte
            tmp_pixel = unpack('>B', tmp_pixel)[0]
            x[i][j] = tmp_pixel
        tmp_label = labels.read(1)
        y[i] = unpack('>B', tmp_label)[0]

    images.close()
    labels.close()
    return (x, y)

def displaychar(image):
    plt.imshow(np.reshape(image, (28,28)), cmap=plt.cm.gray)
    plt.axis('off')
    plt.show()

x,y = loadmnist('train-images-idx3-ubyte', 'train-labels-idx1-ubyte')
#displaychar(x[1])
#print(y[1])
trainset = x[0:50000]
validset = x[50000:]
trainlabel = y[0:50000]
validlabel = y[50000:]

pi = [0] * 10
for index in range(50000):
    pi[trainlabel[index]] += 1
for index in range(10):
    pi[index] = math.log(pi[index]/float(50000))

trainlabel0 = []
trainlabel1 = []
trainlabel2 = []
trainlabel3 = []
trainlabel4 = []
trainlabel5 = []
trainlabel6 = []
trainlabel7 = []
trainlabel8 = []
trainlabel9 = []

for index in range(50000):
    if trainlabel[index] == 0:
        trainlabel0.append(trainset[index])
    elif trainlabel[index] == 1:
        trainlabel1.append(trainset[index])
    elif trainlabel[index] == 2:
        trainlabel2.append(trainset[index])
    elif trainlabel[index] == 3:
        trainlabel3.append(trainset[index])
    elif trainlabel[index] == 4:
        trainlabel4.append(trainset[index])
    elif trainlabel[index] == 5:
        trainlabel5.append(trainset[index])
    elif trainlabel[index] == 6:
        trainlabel6.append(trainset[index])
    elif trainlabel[index] == 7:
        trainlabel7.append(trainset[index])
    elif trainlabel[index] == 8:
        trainlabel8.append(trainset[index])
    else:
        trainlabel9.append(trainset[index])

mean0 = np.mean(trainlabel0, axis=0)
mean1 = np.mean(trainlabel1, axis=0)
mean2 = np.mean(trainlabel2, axis=0)
mean3 = np.mean(trainlabel3, axis=0)
mean4 = np.mean(trainlabel4, axis=0)
mean5 = np.mean(trainlabel5, axis=0)
mean6 = np.mean(trainlabel6, axis=0)
mean7 = np.mean(trainlabel7, axis=0)
mean8 = np.mean(trainlabel8, axis=0)
mean9 = np.mean(trainlabel9, axis=0)

cov0 = np.cov(trainlabel0,rowvar=False)
cov1 = np.cov(trainlabel1,rowvar=False)
cov2 = np.cov(trainlabel2,rowvar=False)
cov3 = np.cov(trainlabel3,rowvar=False)
cov4 = np.cov(trainlabel4,rowvar=False)
cov5 = np.cov(trainlabel5,rowvar=False)
cov6 = np.cov(trainlabel6,rowvar=False)
cov7 = np.cov(trainlabel7,rowvar=False)
cov8 = np.cov(trainlabel8,rowvar=False)
cov9 = np.cov(trainlabel9,rowvar=False)

'''
logprob0 = scipy.stats.multivariate_normal.logpdf(validset,mean0,cov0+3050*np.identity(784))
logprob1 = scipy.stats.multivariate_normal.logpdf(validset,mean1,cov1+3050*np.identity(784))
logprob2 = scipy.stats.multivariate_normal.logpdf(validset,mean2,cov2+3050*np.identity(784))
logprob3 = scipy.stats.multivariate_normal.logpdf(validset,mean3,cov3+3050*np.identity(784))
logprob4 = scipy.stats.multivariate_normal.logpdf(validset,mean4,cov4+3050*np.identity(784))
logprob5 = scipy.stats.multivariate_normal.logpdf(validset,mean5,cov5+3050*np.identity(784))
logprob6 = scipy.stats.multivariate_normal.logpdf(validset,mean6,cov6+3050*np.identity(784))
logprob7 = scipy.stats.multivariate_normal.logpdf(validset,mean7,cov7+3050*np.identity(784))
logprob8 = scipy.stats.multivariate_normal.logpdf(validset,mean8,cov8+3050*np.identity(784))
logprob9 = scipy.stats.multivariate_normal.logpdf(validset,mean9,cov9+3050*np.identity(784))

add0 = [x+pi[0] for x in logprob0]
add1 = [x+pi[1] for x in logprob1]
add2 = [x+pi[2] for x in logprob2]
add3 = [x+pi[3] for x in logprob3]
add4 = [x+pi[4] for x in logprob4]
add5 = [x+pi[5] for x in logprob5]
add6 = [x+pi[6] for x in logprob6]
add7 = [x+pi[7] for x in logprob7]
add8 = [x+pi[8] for x in logprob8]
add9 = [x+pi[9] for x in logprob9]

predict = [0] * 10000
for i in range(10000):
    maxvalue = max(add0[i],add1[i],add2[i],add3[i],add4[i],add5[i],add6[i],add7[i],add8[i],add9[i])
    if maxvalue == add0[i]:
        predict[i] = 0
    elif maxvalue == add1[i]:
        predict[i] = 1
    elif maxvalue == add2[i]:
        predict[i] = 2
    elif maxvalue == add3[i]:
        predict[i] = 3
    elif maxvalue == add4[i]:
        predict[i] = 4
    elif maxvalue == add5[i]:
        predict[i] = 5
    elif maxvalue == add6[i]:
        predict[i] = 6
    elif maxvalue == add7[i]:
        predict[i] = 7
    elif maxvalue == add8[i]:
        predict[i] = 8
    else:
        predict[i] = 9

wrong = 0
for index in range(10000):
    if(predict[index] != validlabel[index]):
        wrong = wrong + 1

print(wrong)
'''

testx,testy = loadmnist('t10k-images-idx3-ubyte', 't10k-labels-idx1-ubyte')

logprob0 = scipy.stats.multivariate_normal.logpdf(testx,mean0,cov0+3050*np.identity(784))
logprob1 = scipy.stats.multivariate_normal.logpdf(testx,mean1,cov1+3050*np.identity(784))
logprob2 = scipy.stats.multivariate_normal.logpdf(testx,mean2,cov2+3050*np.identity(784))
logprob3 = scipy.stats.multivariate_normal.logpdf(testx,mean3,cov3+3050*np.identity(784))
logprob4 = scipy.stats.multivariate_normal.logpdf(testx,mean4,cov4+3050*np.identity(784))
logprob5 = scipy.stats.multivariate_normal.logpdf(testx,mean5,cov5+3050*np.identity(784))
logprob6 = scipy.stats.multivariate_normal.logpdf(testx,mean6,cov6+3050*np.identity(784))
logprob7 = scipy.stats.multivariate_normal.logpdf(testx,mean7,cov7+3050*np.identity(784))
logprob8 = scipy.stats.multivariate_normal.logpdf(testx,mean8,cov8+3050*np.identity(784))
logprob9 = scipy.stats.multivariate_normal.logpdf(testx,mean9,cov9+3050*np.identity(784))

add0 = [x+pi[0] for x in logprob0]
add1 = [x+pi[1] for x in logprob1]
add2 = [x+pi[2] for x in logprob2]
add3 = [x+pi[3] for x in logprob3]
add4 = [x+pi[4] for x in logprob4]
add5 = [x+pi[5] for x in logprob5]
add6 = [x+pi[6] for x in logprob6]
add7 = [x+pi[7] for x in logprob7]
add8 = [x+pi[8] for x in logprob8]
add9 = [x+pi[9] for x in logprob9]

predict = [0] * 10000
for i in range(10000):
    maxvalue = max(add0[i],add1[i],add2[i],add3[i],add4[i],add5[i],add6[i],add7[i],add8[i],add9[i])
    if maxvalue == add0[i]:
        predict[i] = 0
    elif maxvalue == add1[i]:
        predict[i] = 1
    elif maxvalue == add2[i]:
        predict[i] = 2
    elif maxvalue == add3[i]:
        predict[i] = 3
    elif maxvalue == add4[i]:
        predict[i] = 4
    elif maxvalue == add5[i]:
        predict[i] = 5
    elif maxvalue == add6[i]:
        predict[i] = 6
    elif maxvalue == add7[i]:
        predict[i] = 7
    elif maxvalue == add8[i]:
        predict[i] = 8
    else:
        predict[i] = 9

wrong = 0
for index in range(10000):
    if(predict[index] != testy[index]):
        print(index)
        wrong = wrong + 1

print(wrong)
```
